```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()

data= df[['D_Score', 'E_Score',
       'Penalty']].dropna()
X=data.drop(["E_Score"], axis=1)
y=data["E_Score"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4,
)

model.fit(X_train, y_train)
pd.DataFrame(model.coef_, X.columns, columns = ['Coeff'])
print(model.score(X_test, y_test))

predictions = model.predict(X_test)
plt.scatter(y_test, predictions)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder

def preprocess_help(df):
    standard_numerical_transformer = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy="constant", fill_value=0)),
    ])
    categorical_transformer = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy="constant", fill_value="missing")), 
        ("onehot", OneHotEncoder(handle_unknown="ignore", sparse=False))
    ])
    categorical_columns = df.select_dtypes(include=["object", "category"]).columns
    numerical_columns = df.select_dtypes(include=["int64", "float64"]).columns

    preprocessor = ColumnTransformer(
        transformers=[
            ("num", standard_numerical_transformer, numerical_columns), 
            ("cat", categorical_transformer, categorical_columns)
        ]
    )
    return preprocessor

x = df.drop(["Date", "FirstName", "LastName", "Rank", "E_Score", "Penalty", "Score"], axis=1)
y = df["D_Score"] 

preprocessor = preprocess_help(x)
pipeline1 = Pipeline(steps=[
    ("preprocessor", preprocessor)
])
x_preprocessed = pipeline1.fit_transform(x)

x_preprocessed.shape
```
