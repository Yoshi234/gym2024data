```python
import pandas as pd 

data_path = "../../../cleandata/data_2022_2023.csv"

comp2022_2023 = pd.read_csv(data_path)
```

```python
comp2022_2023.head()
```

```python
import numpy as np
'''
pseudo code for cleaning data

create new column: competitor with all NA values
for row in data frame:
    lowercase the last name and first name data inputs
    create a new entry for competitor - "{}, {}".format(LastName, FirstName)
    check the sequence alignment against the existing competitor values (if there are none, skip this step)
        if the similarity has a score of 2 or less, then consider them to be the same
'''

comp2022_2023["Competitor"] = np.nan

def concat_names(first, last):
    return "{} {}".format(first, last)

# concatenate the names after lowercasing them
comp2022_2023["LastName"] = comp2022_2023["LastName"].str.lower()
comp2022_2023["FirstName"] = comp2022_2023["FirstName"].str.lower()
comp2022_2023["Competitor"] = comp2022_2023.apply(lambda x: concat_names(x["FirstName"], x["LastName"]), axis=1)
comp2022_2023.head()
```

```python
def set_competition_name(date, comp, loc):
    ''' 
    take the date, competition, and location and concatenate together to get an overall 
    competition name value - drop the extra values from the data frame
    '''
    # new_date <tuple> = (month, year)
    new_date = (date.split(" ")[-2], date.split(" ")[-1])
    new_loc = loc.split(",")[-1].strip()

    return "{}-{} {} {}".format(new_date[0], new_date[1], comp, new_loc)
    

comp2022_2023["Format Competition"] = comp2022_2023.apply(lambda x: set_competition_name(x["Date"], x["Competition"], x["Location"]), axis=1)
comp2022_2023.head()
comp2022_2023.drop("competitor", axis=1)
```
